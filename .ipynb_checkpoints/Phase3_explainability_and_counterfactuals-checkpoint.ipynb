```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import warnings
warnings.filterwarnings('ignore')

# SHAP for explainability
import shap

# DICE-ML for counterfactuals
import dice_ml
from dice_ml import Dice

# Load models
from sklearn.ensemble import RandomForestClassifier

# Set random seed
np.random.seed(42)

# Plotting settings
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 6)

# Initialize SHAP's JS for notebook visualization
shap.initjs()

print(" All libraries imported successfully!")
```

```python
print("Loading data and trained model...")

# Load processed data
X_train = pd.read_csv('/Users/kashishpatel/Desktop/customer-churn-project/data/processed/X_train.csv')
X_test = pd.read_csv('/Users/kashishpatel/Desktop/customer-churn-project/data/processed/X_test.csv')
y_train = pd.read_csv('/Users/kashishpatel/Desktop/customer-churn-project/data/processed/y_train.csv').values.ravel()
y_test = pd.read_csv('/Users/kashishpatel/Desktop/customer-churn-project/data/processed/y_test.csv').values.ravel()

# Load the best model
with open('/Users/kashishpatel/Desktop/customer-churn-project/models/best_model.pkl', 'rb') as f:
    best_model = pickle.load(f)

# Load model metadata
with open('/Users/kashishpatel/Desktop/customer-churn-project/models/model_metadata.pkl', 'rb') as f:
    model_metadata = pickle.load(f)

print(" Data and model loaded!")
print(f"\nBest Model: {model_metadata['model_name']}")
print(f"ROC AUC: {model_metadata['metrics']['roc_auc']:.4f}")
print(f"\nFeatures: {len(X_train.columns)}")
print(f"Training samples: {X_train.shape[0]:,}")
print(f"Test samples: {X_test.shape[0]:,}")
```

```python
print("="*80)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*80)

# Get feature importance from Random Forest
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': best_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\n TOP 20 MOST IMPORTANT FEATURES:")
print(feature_importance.head(20).to_string(index=False))

# Visualize top 20 features
plt.figure(figsize=(12, 8))
top_20 = feature_importance.head(20)
plt.barh(range(len(top_20)), top_20['Importance'].values, color='steelblue')
plt.yticks(range(len(top_20)), top_20['Feature'].values)
plt.xlabel('Importance Score', fontsize=12)
plt.title('Top 20 Most Important Features (Random Forest)', fontsize=16, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('/Users/kashishpatel/Desktop/customer-churn-project/models/feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n Feature importance analysis complete!")
```

```python
print("="*80)
print("SHAP ANALYSIS - GLOBAL INTERPRETABILITY")
print("="*80)
print("SHAP = SHapley Additive exPlanations")
print("Shows how each feature contributes to predictions\n")

# Create SHAP explainer
# For Random Forest, we use TreeExplainer (fast and accurate)
print("Creating SHAP explainer... (this may take a minute)")
explainer = shap.TreeExplainer(best_model)

# Calculate SHAP values for test set (using a sample for speed)
# For production, you'd use the full test set
sample_size = min(500, len(X_test))
X_test_sample = X_test.sample(n=sample_size, random_state=42)

print(f"Calculating SHAP values for {sample_size} test samples...")
shap_values = explainer.shap_values(X_test_sample)

# For binary classification, shap_values is a list [class_0, class_1]
# We want class_1 (churn = 1)
if isinstance(shap_values, list):
    shap_values_churn = shap_values[1]
else:
    shap_values_churn = shap_values

print(" SHAP values calculated!")
```

```python
print("Creating SHAP summary plot...")

# Summary plot showing feature importance and impact
plt.figure(figsize=(12, 8))
shap.summary_plot(shap_values_churn, X_test_sample, max_display=20, show=False)
plt.title('SHAP Summary Plot - Feature Impact on Churn Prediction', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('/Users/kashishpatel/Desktop/customer-churn-project/models/shap_summary.png', dpi=300, bbox_inches='tight')
plt.show()

print(" SHAP summary plot created!")
print("\n How to read this plot:")
print("   Y-axis: Features ranked by importance")
print("   X-axis: SHAP value (impact on prediction)")
print("   Color: Feature value (red = high, blue = low)")
print("   Each dot is one customer")
print("   Dots to the right increase churn probability")
print("   Dots to the left decrease churn probability")
```

```python
print("Creating SHAP feature importance bar plot...")

# Bar plot showing mean absolute SHAP values
plt.figure(figsize=(12, 8))
shap.summary_plot(shap_values_churn, X_test_sample, plot_type='bar', max_display=20, show=False)
plt.title('SHAP Feature Importance - Top 20 Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/Users/kashishpatel/Desktop/customer-churn-project/models/shap_importance.png', dpi=300, bbox_inches='tight')
plt.show()

print(" SHAP importance plot created!")
```

```python
print("="*80)
print("SHAP ANALYSIS - LOCAL INTERPRETABILITY")
print("="*80)
print("Explaining individual customer predictions\n")

# Get predictions for test sample
y_pred_proba_sample = best_model.predict_proba(X_test_sample)[:, 1]

# Find a high-risk customer (high churn probability)
high_risk_idx = y_pred_proba_sample.argmax()
high_risk_customer = X_test_sample.iloc[high_risk_idx]
high_risk_prob = y_pred_proba_sample[high_risk_idx]

print(f"HIGH-RISK CUSTOMER ANALYSIS:")
print(f"Churn Probability: {high_risk_prob:.1%}")
print(f"Actual Churn: {'Yes' if y_test[X_test_sample.index[high_risk_idx]] == 1 else 'No'}")

# Create waterfall plot for this customer
print("\nGenerating SHAP waterfall plot...")

# Get base value (ensure it's a scalar)
if isinstance(explainer.expected_value, list):
    base_value = explainer.expected_value[1]
else:
    base_value = explainer.expected_value

# Convert to scalar if it's an array
if hasattr(base_value, '__len__'):
    base_value = float(base_value[0]) if len(base_value) > 0 else float(base_value)
else:
    base_value = float(base_value)

# Get SHAP values for this specific customer (ensure it's 1D)
customer_shap_values = shap_values_churn[high_risk_idx]
if len(customer_shap_values.shape) > 1:
    customer_shap_values = customer_shap_values.flatten()

# Create explanation object
shap_explanation = shap.Explanation(
    values=customer_shap_values,
    base_values=base_value,
    data=X_test_sample.iloc[high_risk_idx].values,
    feature_names=X_test_sample.columns.tolist()
)

plt.figure(figsize=(12, 8))
shap.plots.waterfall(shap_explanation, max_display=15, show=False)
plt.title(f'SHAP Waterfall - High Risk Customer (Churn Prob: {high_risk_prob:.1%})', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('/Users/kashishpatel/Desktop/customer-churn-project/models/shap_waterfall_high_risk.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Waterfall plot created!")
print("\n How to read this plot:")
print("   Shows how each feature pushes the prediction higher or lower")
print("   Red bars push prediction toward churn")
print("   Blue bars push prediction away from churn")
print("   Values shown are the actual feature values for this customer")
```

```python
print("Analyzing a LOW-RISK customer...")

# Find a low-risk customer
low_risk_idx = y_pred_proba_sample.argmin()
low_risk_customer = X_test_sample.iloc[low_risk_idx]
low_risk_prob = y_pred_proba_sample[low_risk_idx]

print(f"\nLOW-RISK CUSTOMER ANALYSIS:")
print(f"Churn Probability: {low_risk_prob:.1%}")
print(f"Actual Churn: {'Yes' if y_test[X_test_sample.index[low_risk_idx]] == 1 else 'No'}")

# Show top features contributing to low risk
print("\nTop 10 features pushing AWAY from churn:")
print("-" * 60)

# Get SHAP values for this customer
customer_shap_low = shap_values_churn[low_risk_idx]
feature_names = X_test_sample.columns.tolist()

# Create dataframe of feature contributions
contributions_low = pd.DataFrame({
    'Feature': feature_names,
    'Value': low_risk_customer.values,
    'SHAP_Value': customer_shap_low
}).sort_values('SHAP_Value', ascending=True)

print(contributions_low.head(10).to_string(index=False))

# Create force plot
print("\nGenerating SHAP force plot...")

# Get base value
if isinstance(explainer.expected_value, list):
    base_value = explainer.expected_value[1]
else:
    base_value = explainer.expected_value

# Create force plot (saved as matplotlib)
shap.force_plot(
    base_value,
    customer_shap_low,
    low_risk_customer,
    feature_names=feature_names,
    matplotlib=True,
    show=False
)
plt.title(f'SHAP Force Plot - Low Risk Customer (Churn Prob: {low_risk_prob:.1%})', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('/Users/kashishpatel/Desktop/customer-churn-project/models/shap_force_low_risk.png', 
            dpi=300, bbox_inches='tight')
plt.show()

print(" Low-risk customer analysis complete!")
```

```python
print("="*80)
print("COUNTERFACTUAL EXPLANATIONS WITH DICE-ML")
print("="*80)
print("Generating 'What-If' scenarios to reduce churn\n")

# Prepare data for DICE
# DICE needs the full dataset with target variable
df_train = X_train.copy()
df_train['Churn'] = y_train

df_test_dice = X_test.copy()
df_test_dice['Churn'] = y_test

# Create DICE data object
dice_data = dice_ml.Data(
    dataframe=df_train,
    continuous_features=X_train.select_dtypes(include=[np.number]).columns.tolist(),
    outcome_name='Churn'
)

# Create DICE model object
dice_model = dice_ml.Model(
    model=best_model,
    backend='sklearn'
)

# Create DICE explainer
print("Creating DICE explainer...")
dice_exp = Dice(dice_data, dice_model, method='random')

print(" DICE explainer created!")
```

```python
print("Generating counterfactual explanations for HIGH-RISK customer...\n")

# Select the high-risk customer we analyzed earlier
high_risk_customer_df = X_test_sample.iloc[[high_risk_idx]]

# Generate counterfactuals
# We want to find what changes would flip prediction to "No Churn"
print("Finding alternative scenarios that would prevent churn...")
cf_explanations = dice_exp.generate_counterfactuals(
    high_risk_customer_df,
    total_CFs=5,  # Generate 5 counterfactual scenarios
    desired_class="opposite"  # We want the opposite prediction
)

# Visualize counterfactuals
print("\n COUNTERFACTUAL SCENARIOS:")
print("="*80)
cf_explanations.visualize_as_dataframe(show_only_changes=True)
```

```python
print("\n" + "="*80)
print("ACTIONABLE RECOMMENDATIONS FROM COUNTERFACTUALS")
print("="*80)

# Get the counterfactuals as a dataframe
cf_df = cf_explanations.cf_examples_list[0].final_cfs_df

# Compare original vs counterfactuals
print(f"\n ORIGINAL CUSTOMER (Churn Probability: {high_risk_prob:.1%}):")
print("-" * 80)

# Show key features for original customer
key_features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Contract', 
                'TotalServices', 'IsMonthToMonth']
# Filter to features that exist
key_features = [f for f in key_features if f in high_risk_customer.index]

for feature in key_features:
    if feature in high_risk_customer.index:
        print(f"  {feature}: {high_risk_customer[feature]}")

print(f"\n RECOMMENDED CHANGES TO PREVENT CHURN:")
print("-" * 80)

# Extract changes from counterfactuals
if len(cf_df) > 0:
    # Get the first (best) counterfactual
    cf_best = cf_df.iloc[0]
    
    changes = []
    for feature in high_risk_customer.index:
        if feature in cf_best.index:
            original_val = high_risk_customer[feature]
            cf_val = cf_best[feature]
            
            if original_val != cf_val:
                change = {
                    'Feature': feature,
                    'Original': original_val,
                    'Recommended': cf_val,
                    'Change': cf_val - original_val if isinstance(original_val, (int, float)) else 'Change needed'
                }
                changes.append(change)
    
    # Print top changes
    print("\nTop recommended interventions:")
    for i, change in enumerate(changes[:5], 1):
        print(f"\n{i}. {change['Feature']}:")
        print(f"   Current: {change['Original']}")
        print(f"   Recommended: {change['Recommended']}")
        if isinstance(change['Change'], (int, float)):
            direction = "increase" if change['Change'] > 0 else "decrease"
            print(f"   Action: {direction} by {abs(change['Change']):.2f}")

print("\n BUSINESS INTERPRETATION:")
print("-" * 80)
print("These changes represent actions the company can take:")
print("   Offer contract upgrades (Month-to-month  Long-term)")
print("   Provide service bundles to increase engagement")
print("   Offer loyalty discounts on monthly charges")
print("   Improve customer tenure through retention programs")

print("\n Counterfactual analysis complete!")
```

```python
print("Creating comprehensive counterfactual report...\n")

# Function to generate counterfactual recommendations
def generate_recommendations(customer_data, customer_idx):
    """Generate counterfactual recommendations for a customer"""
    
    # Get customer info
    customer_df = customer_data.iloc[[customer_idx]]
    churn_prob = best_model.predict_proba(customer_df)[0, 1]
    
    # Generate counterfactuals
    try:
        cf_exp = dice_exp.generate_counterfactuals(
            customer_df,
            total_CFs=3,
            desired_class="opposite"
        )
        
        cf_df = cf_exp.cf_examples_list[0].final_cfs_df
        
        # Extract changes
        changes = []
        if len(cf_df) > 0:
            cf_best = cf_df.iloc[0]
            original = customer_df.iloc[0]
            
            for feature in original.index:
                if feature in cf_best.index and original[feature] != cf_best[feature]:
                    changes.append({
                        'feature': feature,
                        'from': original[feature],
                        'to': cf_best[feature]
                    })
        
        return {
            'churn_probability': churn_prob,
            'recommendations': changes[:5]  # Top 5 changes
        }
    except:
        return {
            'churn_probability': churn_prob,
            'recommendations': []
        }

# Generate recommendations for top 10 high-risk customers
print("Analyzing top 10 high-risk customers...\n")

high_risk_indices = y_pred_proba_sample.argsort()[-10:][::-1]

recommendations_report = []

for idx in high_risk_indices:
    result = generate_recommendations(X_test_sample, idx)
    recommendations_report.append({
        'Customer_Index': X_test_sample.index[idx],
        'Churn_Probability': result['churn_probability'],
        'Top_Recommendation': result['recommendations'][0]['feature'] if result['recommendations'] else 'N/A'
    })

# Create summary dataframe
report_df = pd.DataFrame(recommendations_report)

print("="*80)
print("TOP 10 HIGH-RISK CUSTOMERS - INTERVENTION PRIORITIES")
print("="*80)
print(report_df.to_string(index=False))

# Save report
report_df.to_csv('/Users/kashishpatel/Desktop/customer-churn-project/models/high_risk_recommendations.csv', index=False)

print("\n Report saved: models/high_risk_recommendations.csv")
```

```python
print("="*80)
print("BUSINESS IMPACT ANALYSIS")
print("="*80)

# Assumptions for business impact
avg_customer_lifetime_value = 2000  # dollars
cost_per_intervention = 50  # dollars (e.g., discount, customer service)
intervention_success_rate = 0.30  # 30% of interventions prevent churn

# Identify high-risk customers (probability > 0.7)
high_risk_threshold = 0.7
y_pred_proba_test = best_model.predict_proba(X_test)[:, 1]
high_risk_customers = (y_pred_proba_test > high_risk_threshold).sum()

print(f"\n CUSTOMER SEGMENTATION:")
print(f"Total test customers: {len(y_test):,}")
print(f"High-risk customers (>70% churn probability): {high_risk_customers:,}")
print(f"Percentage: {(high_risk_customers/len(y_test)*100):.1f}%")

# Calculate potential savings
potential_saves = high_risk_customers * intervention_success_rate
intervention_cost = high_risk_customers * cost_per_intervention
revenue_saved = potential_saves * avg_customer_lifetime_value
net_benefit = revenue_saved - intervention_cost

print(f"\n FINANCIAL IMPACT (Test Set Projection):")
print(f"Customers we could save: {potential_saves:.0f}")
print(f"Cost of interventions: ${intervention_cost:,.0f}")
print(f"Revenue saved: ${revenue_saved:,.0f}")
print(f"Net benefit: ${net_benefit:,.0f}")
print(f"ROI: {(net_benefit/intervention_cost*100):.0f}%")

# Extrapolate to full customer base
full_customer_base = 100000  # example
scaling_factor = full_customer_base / len(y_test)

print(f"\n PROJECTED ANNUAL IMPACT (for {full_customer_base:,} customers):")
print(f"High-risk customers: {int(high_risk_customers * scaling_factor):,}")
print(f"Customers saved: {int(potential_saves * scaling_factor):,}")
print(f"Total intervention cost: ${int(intervention_cost * scaling_factor):,}")
print(f"Total revenue saved: ${int(revenue_saved * scaling_factor):,}")
print(f"Total net benefit: ${int(net_benefit * scaling_factor):,}")

print("\n Business impact analysis complete!")
```
